```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sst.jpg', 0)


#Plot the original image
plt.subplot(1, 4, 1)
plt.title("Original")
plt.imshow(img)


#Gaussian Blur
Gaussian = cv.GaussianBlur(img, (7, 7), 0)
#cv.imshow('Gaussian Blurring', Gaussian)
cv.waitKey(0)
#Plot the Gaussian image
plt.subplot(1, 4, 2)
plt.title("Gaussian")
plt.imshow(Gaussian)



#Median Blur
Median = cv.medianBlur(img, 5)
#cv.imshow('Median Blurring', Median)
cv.waitKey(0)
#Plot the Median image
plt.subplot(1, 4, 3)
plt.title("Median")
plt.imshow(Median)

           
#Bilateral Blur
Bilateral = cv.bilateralFilter(img, 9, 75, 75)
#cv.imshow('Bilateral Blurring', Bilateral)
cv.waitKey(0)
#Plot the Bilateral image
plt.subplot(1, 4, 4)
plt.title("Bilateral")
plt.imshow(Bilateral)

cv.waitKey(0)
cv.destroyAllWindows()
```
