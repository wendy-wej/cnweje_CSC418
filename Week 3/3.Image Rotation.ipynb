```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sst.jpg', 0)
rows, cols = img.shape

#Plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)


M = cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6)
rotated_img = cv.warpAffine(img, M, (cols, rows))

#Plot the original image
plt.subplot(1,2,2)
plt.title("Rotated Image")
plt.imshow(rotated_img)

cv.imshow("img", rotated_img)
cv.imwrite("img/rotated_out.jpg", rotated_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
